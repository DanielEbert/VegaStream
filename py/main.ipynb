```python
import altair as alt
```

```python
charta = alt.Chart({'name': 'source'}).mark_point().encode(
    x='x:Q',
    y='y:Q'
).interactive()

chartb = alt.Chart({'name': 'sourceb'}).mark_point(color='red').encode(
    x='a:Q',
    y='b:Q'
).interactive()

comb = charta + chartb

print(comb.to_json())
```
